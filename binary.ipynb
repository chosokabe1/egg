```python
import cv2
import numpy as np
from PIL import Image
from IPython import display
```

```python
def imshow(img, format=".jpg", **kwargs):
    img = cv2.imencode(format, img)[1]
    img = display.Image(img, **kwargs)
    display.display(img)
```

```python
def a_image_max_height_width(im):
    h, w= im.shape
    if h >= w:
        max = h
    else:
        max = w
    
    return max
```

```python
def make_square(im,MAX_HW):
    h, w= im.shape

    out_im = np.zeros((MAX_HW,MAX_HW))
    print(im)
    out_im[0:h, 0:w] = im
    
    cv2.imwrite("./a.jpg", out_im)

    return out_im
```

```python
img = cv2.imread('../0.jpg',0)
img = cv2.bitwise_not(img)
imgCopy = np.uint8(img)
img_canny = cv2.Canny(imgCopy, 40, 40)
imshow(img)
imshow(img_canny)

```

```python
kernel1 = np.ones((5,5),np.uint8)
kernel2 = np.ones((35,35),np.uint8)
kernel3 = np.ones((135,135),np.uint8)
```

```python
closing2 = cv2.morphologyEx(img_canny, cv2.MORPH_CLOSE, kernel2)
closing2_opening2 = cv2.morphologyEx(closing2, cv2.MORPH_OPEN, kernel2)
imshow(closing2_opening2)
cv2.imwrite("../0binary.jpg", closing2_opening2)
```
