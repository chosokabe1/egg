```python
import cv2
import numpy as np
from PIL import Image
from IPython import display
```

```python
def imshow(img, format=".jpg", **kwargs):
    img = cv2.imencode(format, img)[1]
    img = display.Image(img, **kwargs)
    display.display(img)
```

```python
def a_image_max_height_width(im):
    h, w= im.shape
    if h >= w:
        max = h
    else:
        max = w
    
    return max
```

```python
def make_square(im,MAX_HW):
    h, w= im.shape

    out_im = np.zeros((MAX_HW,MAX_HW))
    print(im)
    out_im[0:h, 0:w] = im
    
    cv2.imwrite("./a.jpg", out_im)

    return out_im
```
